## Thresholding the image to binary

This might be optional. We threshold the image to keep only two values. The parameter of interest here is `threshold` which defines the values above which we define the pixel to be on.

```python
def threshold_images(img_array, threshold):
    results = []
    for img in img_array:
        thresholded_img = img.copy()  # we need to copy to avoid in place
        thresholded_img[img > threshold] = 1
        thresholded_img[img <= threshold] = 0
        
        results.append(thresholded_img)
    
    return results
    
```

```python
threshold = 114
thresholded_images = threshold_images(cropped_array, threshold)
```

```python
show(thresholded_images)
```

## Transforming the images to Fourier space

```python
def fft_images(img_array):
    results = []
    for img in img_array:
        ffted = np.fft.fftn(img)
        results.append(np.log(np.abs(np.fft.fftshift(ffted))**2))
    return results
```

```python
ffted_images = fft_images(thresholded_images)
```

```python
show(ffted_images)
```

## Zooming in on images

```python
def zoom_images(img_array, zoom_factor):
    results = []
    for img in img_array:
        nx, ny = img.shape
        img_center = (nx//2, ny//2)
        results.append(img[img_center[0]-nx//zoom_factor/2:img_center[0]+nx//zoom_factor/2,
                          img_center[1]-ny//zoom_factor/2:img_center[1]+ny//zoom_factor/2])
    return results
        
        
```

```python
zoomed_images = zoom_images(ffted_images, 5)
```
